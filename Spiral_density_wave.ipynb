## Use rotating ellipse to generate a spiral pattern

Reference:

[Spiral Structure Viewed as a Density Wave](http://adsabs.harvard.edu/cgi-bin/bib_query?1973PASAu...2..174K) by A. J. Kalnajs 1973

[Six Decades of Spiral Density Wave Theory](https://www.annualreviews.org/doi/full/10.1146/annurev-astro-081915-023426) by Frank Shu 2016

```python
from matplotlib.patches import Ellipse
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10,10))

mean = [0, 0]
width = 30
height = 25
c=0.7

n = 50
r = 1
for angle in np.linspace(0,180,n):
    
    ell = mpl.patches.Ellipse(xy=mean, width=r, height=r*c, angle = 180+angle, fill=False)
    ax.add_patch(ell)
    r = r+n/20.0
    
ax.set_aspect('equal')
ax.autoscale()
plt.show()
```
